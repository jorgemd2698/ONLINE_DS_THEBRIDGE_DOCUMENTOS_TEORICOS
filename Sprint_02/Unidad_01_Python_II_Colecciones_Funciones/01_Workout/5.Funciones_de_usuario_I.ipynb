### Definicin y sintaxis

Se trata de bloques de cdigo que encapsulan una serie de operaciones. Se usan para modular nuestro cdigo y evitar escribir cdigo de ms.  

Mediante las **funciones** podemos encapsular cdigo en formato entrada/salida. Por lo que si tienes un cdigo repetitivo, que depende de ciertos inputs, las funciones pueden ser una buena solucin.



Es una manera de agrupar conjuntos de operaciones en mdulos. **Cundo usarlas?** Cuando tengamos varias operaciones que ejecutamos repetidamente en distintas partes del cdigo. En ese caso, encapsulamos las operaciones en una funcin, y cada vez que haya que realizar tal operativa, llamamos a la funcin, y en una sola lnea de cdigo tenemos ejecutada esas operaciones.

Hasta ahora hemos estado utilizando funciones *built-in*, para operaciones sencillas como `len()`, `sum()` o `max()`. En este Notebook aprenderas a crear tus propias funciones.

La sintaxis es:
> ```Python
> def nombre_funcion(entrada):
>    operaciones varias
>    return output
> ```

Fjate que sigue la **sintaxis de lnea** vista en Notebooks anteriores (tabulacin detrs de los dos puntos, por ejemplo).  

Adems, todo lo que va despus del `return` es ignorado, puesto que es la salida. En el `return` acaba la funcin. Ahora bien, eso no quiere decir que haya un nico return. Si introducimos una sentencia `if/else`, podremos poner returns diferentes dependiendo de qu condicin se cumpla.

Vamos a crear nuestra primera funcin

````python
# Vamos a crear un conversor de km a millas
def conversor_km_millas(distancia):
    millas = 0.62*distancia
    return millas
````

Al ejecutar el cdigo anterior, no corre nada, simplemente almacenamos en memoria la funcin para usarla posteriormente. **Ahora podremos llamar a la funcin tantas veces como queramos, desde cualquier parte del codigo.**

````python
conversor_km_millas(200)
````

````python
millas = conversor_km_millas(200) #Ojo, no son la variable millas de dentro de la funcin
print(millas)
````

````python
def conversor_km_millas(distancia):
    millas = 0.62*distancia
    return round(millas, 1)  # Redondeamos a 1 decimal
````

````python
millas = conversor_km_millas(2212)
print(millas)
````

Las funciones no tienen por qu llevar argumentos. Eso s, **es obligatorio** poner los parentesis, tanto en la declaracin, como luego al llamar la funcin.

````python
from datetime import datetime
def que_hora_es:
    now = datetime.now().time()
    return now
````

````python
from datetime import datetime
def que_hora_es():
    now = datetime.now().time()
    return now

print(que_hora_es()) #Ojo tambien hay que llamar a la funcin con parntesis
````
