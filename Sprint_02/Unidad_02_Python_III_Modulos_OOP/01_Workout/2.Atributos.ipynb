Son las **caractersticas que definen a los objetos de una clase**. La marca, el color, potencia del coche. Estos son atributos, que se definen de manera genrica en la clase y luego cada objeto *Coche* tendr un valor para cada uno de sus atributos.

Los atributos los definimos tras la declaracin de la clase. Y luego se accede a ellos mediante la sintaxis `objeto.atributo`

Vamos a empezar a definir atributos en los coches.

```python
class Coche():
    puertas = 4
    ruedas = 4
```

Ahora todos los coches que creamos, tendrn 4 puertas y 4 ruedas.

```python
tesla = Coche()
print(tesla.puertas)
print(tesla.ruedas)
```

Tambin podemos modificar los atributos. Esto Python lo hace muy sencillo, los cambiamos directamente reasignando valores. En otros lenguajes de programacin hay que implementar esto mediante mtodos  denominados `getters` y `setters`.

```python
mercedes = Coche()
mercedes.puertas = 3

print(mercedes.puertas)
print(tesla.puertas)
```

```python
#En el ejemplo anterior, cada objeto tiene su propio atributo puertas.
#Si modificamos el atributo puertas de mercedes, no afecta al atributo puertas de tesla
peugeot = Coche()
print(peugeot.puertas)
```
